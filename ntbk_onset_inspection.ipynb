## Grab the data files

```python
# onsets=glob.glob('/projects/niblab/experiments/bbx/data/onsets/output_onsets/*txt')
logfiles=glob.glob('/projects/niblab/experiments/bbx/data/onsets/log_files/bbx_*')
onsets=glob.glob('/projects/niblab/experiments/bbx/data/onsets/trimmed_evs/*.txt')

onsets.sort()
logfiles.sort()
```

```python
data_dict={}
```

```python
for x in logfiles:
    y=x.split("/")[-1].split("_")[1]
    r=x.split("/")[-1].split("_")[2]
    c=x.split("/")[-1].split("_")[3]
    name=y+"_"+r+"_"+c
    
    if y not in data_dict:
        data_dict[y]={}
    
    if 'pre' in c:
        data_dict[y]['pre_%s'%r]='True'
    elif 'post' in c:
        data_dict[y]['post_%s'%r]='True'
    
    else:
        data_dict[y]['file_%s'%r]='True'
```

```python
df=pd.DataFrame(data_dict).T
```
